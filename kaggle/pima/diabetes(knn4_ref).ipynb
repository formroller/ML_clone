#[2.Skweness in statistics](https://www.datasciencecentral.com/profiles/blogs/skewness-in-statistics)
 통계량의 왜도 및 첨도

image.png


문자 그대로 왜도는 <u>'대칭의 결여'</u>를 의미한다. 우리는 주어진 데이터로 그릴 수 있는 곡선의 모양에 대한 아이디어를 얻기 위해 왜도를 연구한다. 

* 분포는 다음과 같은 경우에 치우쳐 있다고 한다. 
 * Mean, median, mode는 서로 다른 점에서 하강한다. 
 * 사분위수는 중위수로부터 등거리에 있지 않는다.
 * 주어진 데이터로 그려진 곡선은 대칭적이지 않고 다른 쪽보다 한쪽으로 더 많이 뻗어져 있다.

분포의 대칭성 결여는 항상 대칭인 정규분포를 참조해 결정된다. 분포가 대칭에서 벗어나면 비대칭 분포가 발생하며 이러한 경우 이 분포를 편향이라고 한다. 왜도는 양수 또는 음수일 수 있으며, 왜도가 없으면 분포가 대칭이 된다.  
 분포의 왜도는 단순히 검사만으로 결정할 수 없다는 것이 중요하다. 평균, 중앙값, 최빈값의 차이를 이해하면 왜도 방향을 제안할 수 있어야 한다.   
 도수 분포의 왜도를 다음과 같이 세가지 다른 모양으로 정의할 수 있다. 

## 2.1 왜도 분포


### 2.1.1 대칭분포
평균, 중앙값, 최빈값이 동일한 값을 갖는다.  
(예 : 키, 체중 등)


image.png

### 2.1.2 양으로 치우친 분포
오른쪽으로 꼬리가 길어 분포의 양수 끝에 극단값이 있음을 나타내며 평균을 오른쪽으로 이동시킨다.  
(예 : 가족 규모, 근로자 임금 등)  
image.png

### 2.1.3 음으로 치우친 분포
이 분포의 유형에서 평균은 음의 방향으로 당겨진다.  
(예 : 겨울 한 달 동안 일 최고 기온)  
image.png


## 2.2 왜도 측정
아래와 같이 Pearson의 왜도 계수를 사용해 간단하게 왜도를 측정할 수 있다.  

$$(Pearson 왜도 계수) \  Skewness(p) = \frac{Mean-Mode}{Standar Deviation}$$
 
 * 우리는 다음과 같이 피어슨의 왜도 계수에 대해 다음과 같은 결정을 내릴 수 있다.
 1. 평균 > 최빈값, 분포가 양의 방향
 2. 평균 = 모드, 분포가 치우치거나 대칭적이지 않다.
 3. 평균 < 최빈값, 분포가 음의 방향

최빈값을 고유하게 정의할 수 없는 경우가 있으므로 그 경우 아래와 같은 방법으로 도출한다.
$$Skewness(p) = 3\frac{Mean-Mode}{Standard Deviation}$$

## 4.1 모델의 Overfitting / Underfiting 이란?
통계와 기계학습에서 데이터를 훈련용/테스트용으로 나누고 테스트 데이터를 예측하기 위해 train 데이터에 모델을 맞춘다. 이때 만들어둔 모델이 답을 과하게 맞추거나, 적게 맞추는 경우를 말한다.

image.png

### 4.1.1 과적합(Overfitting)
우리가 훈련한 모델이 "과하게" 훈련되었고, 훈련 데이터 세트에 너무 가깝게 적합하는 것을 의미한다. 이는 모형이 너무 복잡할 때 발생하며 과적합된 모델은 일반화 할 수 없기 때문에 다른 데이터에 대한 추론이 불가능하다.   
 이 경우 모델은 데이터 내 변수간의 실제 관계 대힌 '잡음'을 학습하거나 설명하게 된다.

### 4.1.2 과소적합(Underfitting)
모형이 불충분한 경우에 교육 데이터에 적합하지 않으므로 데이터의 추세를 놓치게 된다. 과적합과 마찬가지로 예측 능력이 저하되며 일반화 할 수 없기 때문에 다른 데이터에 대한 추론이 불가능하다.


## 4.2 Train / Test Split
전체 데이터를 모델 훈련에 사용하는 훈련용 데이터(train)와 검정용 데이터(test)로 분리하는 것을 의미한다. 

## 4.3 Cross Validation
고정된 test set을 통해 모델을 성능을 확인하고 파라미터를 수정하는 과정을 반복할 경우 결국 test set에만 잘 동작하는 과적합(Overfitting) 모델이 되며 이를 해결하기 위해 교차 검증(Cross Validation)을 사용한다.

image.png

### 4.2.1 K-Folds Cross Validation
* 데이터를 k개의 서로 다른 하위 집합으로 분할한다. 
* k-1개의 하위 세트랄 사용해 데이터를 훈련하고 마지막 세트를 테스트용으로 남겨둔다.
* 각 집합을 기준으로 모델을 평균화한다.
* 테스트 세트와 비교한다

image.png


**장점**
1. 모든 데이터 셋을 평가에 활용할 수 있다. 
 - 평가에 사용되는 데이터 편중 방지.
 - 평가 결과에 따라 좀 더 일반화된 모델 생성 가능.
2. 모든 데이터 셋을 훈련에 활용할 수 있다.
 - 정확도 향상 가능.
 - 데이터 부족으로 인한 underfitting 방지 가능.

**단점**
Iteration 횟수가 많기에 모델 훈련/평가에 시간이 오래 걸린다.

### 4.2.2 Leave One Out Cross Validation (LOOCV)
리브-원-아웃 교차검증   
* leave-p-out 방법에서 p=1 인 경우.
* leave-p-out-cross validation보다 계산 시간에 대한 부담 감소. 
* 검증에 사용되는 test set의 갯수가 적은 만큼 모델 훈련에 사용되는 데이터의 갯수 향상  
    -> 모델 검증에 희생되는 데이터의 갯수가 하나이기 때문에, 나머지 모든 데이터를 모델 훈련에 사용할 수 있는 것이 장점.

## 5.1 ROC 분석이란?
ROC Curve는 (Receiver Operating characteristic Curve) 민감도(Sensitivity)와 1-특이도(Specificity)로 그려지는 곡선을 의미하며 주로 
* 검사도구의 유용성을 판단하거나 검사의 정확도를 평가
* 진단을 위한 도구 개발에서 검사의 기준점(Cut Point)을 설정하는 경우  
등에 사용된다. 

분석을 위해서는 최소한 2개 이상의 변수가 있어야 하고 두 개의 변수 중 하나는 반드시 진단(확진) 결과 또는 검사 결과를 나타내는 <u>이진형 변수</u>여야 한다.

### 5.1.1 개념 설명
ROC Curve에서 가장 핵심적인 개념들이 모두 아래 표에서 파생된다.

image.png


가설 검정의 원리와 크게 차이가 없으며 굳이 비교하자면 **FP**가 **1종 오류(Type Ⅰ Error)**에 해당하고 **FN**이 **2종 오류(Type Ⅱ Error)**에 해당한다.

**위 표에 대해 이해해보자**
* 예측값 : Positive / Negative로 구분하며 <u>사전 예측값</u>이다.
 * Positive, 양성
 * Negative, 음성

* 실제값 :True / False로 구분하며 <u>실제 진단결과</u>이며 예측값과의 일치여부를 나타낸다. 
 * True, 일치
 * False, 불일치

따라서 TP, FN, FP, TN은 다음과 같이 설명될 수 있다.

|분류|내용|명칭|
|----|----|----|
|TP|양성(P)으로 예측했는데 진단이 양성이므로 예측과 진단 **일치**| True Positive (예측(양성) == 진단(양성)|
|FN|음성(N)으로 예측했으나 진단은 양성이므로 **불일치**|False Negative 예측(음성) != 진단(양성)|
|FP|양성(P)으로 예측했으나 진단은 음성이므로 **불일치** |False Positive (예측(양성) != 진단(음성))|
|TN|음성(N)으로 예측했으나 진단이 음성이므로 **일치**| True Negative (예측(음성) == 진단(음성)|


### 5.1.2 민감도(Sensitivity)와 특이도 (Specificity)
ROC 분석에 등장하는 모든 개념들 (민감도, 특이도, 양성 예측도, 정밀도, 음성 예측도, 위양성율, 위음성율, 오분류율, 유병율)은 위 4가지(TP, FN, FP, TN)가 조합되어 계산된다. 하지만 실제 ROC Curve를 그리는데 활용되는 갑은 <u>**민감도**</u>와 <u>**특이도**</u>이고 가장 핵심 개념이다. 

#### 5.1.2.a 민감도 (Sensitivity)
* **실제 양성대 맞춘 양성의 비율**  

<br>

양성으로 예측해서 맞췄거나 음성으로 예측했는데 틀린 경우를 합친 것이 실제 진단 결과가 양성인 것이다. 그리고 여기서 맞춘 것은 양성 **TP**만 해당하므로 이를 비율로 표현한 민감도는 다음과 같이 구해진다.
$$Sensitivity(SE) = \frac{True Positive(TP)}{True Positive(TP) + False Negative(FN)}$$

오판(FN)의 비중이 낮을수록 검사의 정확성(양성을 정확히 예측할 확률)이 증가하기 때문에 <u>민감도는 클수록 좋다.</u>

#### 5.1.2.b 특이도 (Specificity)
* **실제 음성대 맞춘 음성의 비율**  

<br>

실제 음성인 것은 TN과 FP이다. **음성으로 예측해서 맞췄거나 양성으로 예측했는데 틀린 경우를 합친것이 실제 진단 결과가 음성인 경우**에 속한다. 여기서 맞춘 음서은 TN만 해당하므로 이를 비율로 표현한 특이도는 아래와 같이 계산된다.

$$Specificity(SP) = \frac{True Negative(TN)}{True Negative(TN) + False Positive(FP)}$$

<u>특이도도 민감도와 마찬가지로 클수록 좋다</u>  
검사가 양성인 사람을 제대로 판단하는 비율이 크더라도 음성인 사람을 제대로 걸러내지 못하면 검사 전체의 정확성은 떨어질 수 있다.  
따라서 민감도와 마찬가지로 특이도 역시 값이 클수록 음성인 사람을 잘 필터링하는 것이기 때문에 FP의 비중이 낮을수록 좋다.

### 5.1.3 ROC Curve 그리기
* ROC Cureve는 민감도(SE)와 1-특이도(SP)로 그려진다. 

* 왜 x축이 특이도가 아니라 1-특이도일까?
 이는 AUC면적에 대한 설명에서 다루고 우선,1-특이도의 의미인지만 알아두도록 하자. 1-특이도를 수식으로 정리하면 다음과 같이 정리된다.

 $$ 1-Specificity(SP) = 1 - \frac{TN}{TN+FP} = \frac{(TN+FP)-TN}{TN + FP} = \frac{FP}{TN + FP}$$

 그리고 여기서 **FP/TN+FP는 실제로 음성인 것과 양성으로 잘못 예측한 것의 비율로 <u>위양성율(False Positive Rate;FPR)을 의미</u>**한다.

 $$False Positive Rate(FPR) = \frac{False Positive(FP)}{False Positive(FP) + True Negative(TN)} $$

## 5.2 AUC

## 5.2.1 AUC 의미란?
Area Under Curve의 약자로 ROC분석을 통해 그려지는 <u>**'곡선 밑의 면적'**</u>을 의미한다.

1. AUC는 0.5보다 크고 'y=x'선(대각선)을 참조선이라 하는데 이보다 위로 그려져야 하는 이유는?

$$Sensitivity(SF,특이도) = \frac{True Positive(TP)}{True Positive(TP) + False Negative(FN)}$$

$$1-Specificity(SP) = 1-\frac{TN}{TN+TP} = \frac{(TN+FP)-TN}{TN+FP} = \frac{FP}{TN+FP}$$
위에서 1-SP는 다시 위양성률(FPR)이라고 정리했고 다음과 같다.  

$$False Positive Rate(FPR) = \frac{FP}{FP+TN}$$

  - 민감도(TPR) : 양성으로 예측하고 실제도 양성인 경우와 비율
  - 특이도(TNR) : 음성으로 예측하고 실제도 음성인 경우와 비율

- 위양성율(FPR)
 * 1 - 특이도(SP)
 * 실제 음성인 경우를 양성으로 잘못 예측한 경우와 비율

<BR>

2. **정리하자면**
ROC Curve에서   
 * X축 : 음성인 경우를 양성으로 잘못 예측한 비율 (FPR)
 * Y축 : 양성인 경우를 제대로 예측한 비율 (TPR)  
이며, 민감도는 클수록 좋고, 1-특이도(위양성율)은 작을수록 좋다.

만약 민감도와 1-특이도(위양성율)이 1:1로 대응(TPR=FPR)이 된다면 어떻게 될까? 이 경우 ROC Curve는 그래프 참조선상인 y=x의 일차함수에 찍힐 것이다. 이때 AUC 면적은 0.5가 되는데 이 경우의 진단 도구의 정확성은 최악이라고 볼 수 있다. (확률이 50:50이 되므로 진단 도구로써의 가치가 없어진다)
  

따라서 ROC Curve에서 그려지는 곡선 밑 면적(AUC)는 반드시 0.5보다 커야하고 일반적으로 0.7이상 되어야 수용할만한 수준이라고 평가한다.  
-> 즉, AUC 면적이 클수록 해당 검사도구의 정확도가 높다고 볼 수 있다.


|Area Under Curve(AUC)|Evaluation|
|---|---|
|AUC ≥ 0.9 | Excellent|
|0.8 ≤ AUC < 0.9 | Good |
|0.7 ≤ AUC < 0.8 | Fair |
|AUC < 0.7 | Poor |
